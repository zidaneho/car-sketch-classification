```python
from quickdraw import QuickDrawData
import numpy as np
import os
from PIL import Image
```

```python
# Use the categories from your existing list
categories = car_categories + not_car_categories
SAMPLES_PER_CATEGORY_NON_CARS = 2500
SAMPLES_PER_CATEGORY_CARS = len(not_car_categories) * SAMPLES_PER_CATEGORY_NON_CARS // len(car_categories)

# Initialize library with a limit so it doesn't download millions
qd_car = QuickDrawData(max_drawings=SAMPLES_PER_CATEGORY_CARS)
qd_not_car = QuickDrawData(max_drawings=SAMPLES_PER_CATEGORY_NON_CARS)
os.makedirs('data', exist_ok=True)

for category in car_categories:
    save_path = f"data/{category}.npy"
    if not os.path.exists(save_path):
        print(f"Processing {category}...")
        group = qd_car.get_drawing_group(category)
        
        bitmaps = []
        # Draw and resize each drawing to 28x28 to match the QuickDraw .npy format
        SAMPLES_PER_CATEGORY = SAMPLES_PER_CATEGORY_CARS
        for i in range(min(SAMPLES_PER_CATEGORY, group.drawing_count)):
            drawing = group.get_drawing(i)
            # Render to PIL image (default is 255x255 white/black)
            img = drawing.get_image() 
            # Convert to Grayscale ('L') and resize to 28x28
            img_small = img.convert('L').resize((28, 28))
            # Invert colors: QuickDraw library is black-on-white, 
            # but standard ML datasets (and your code) expect white-on-black (0-255)
            data = 255 - np.array(img_small)
            bitmaps.append(data.flatten())
            
        np.save(save_path, np.array(bitmaps))
    else:
        print(f"{category} already exists.")
        
        
for category in not_car_categories:
    save_path = f"data/{category}.npy"
    if not os.path.exists(save_path):
        print(f"Processing {category}...")
        group = qd_not_car.get_drawing_group(category)
        
        bitmaps = []
        # Draw and resize each drawing to 28x28 to match the QuickDraw .npy format
        SAMPLES_PER_CATEGORY = SAMPLES_PER_CATEGORY_NON_CARS
        for i in range(min(SAMPLES_PER_CATEGORY, group.drawing_count)):
            drawing = group.get_drawing(i)
            # Render to PIL image (default is 255x255 white/black)
            img = drawing.get_image() 
            # Convert to Grayscale ('L') and resize to 28x28
            img_small = img.convert('L').resize((28, 28))
            # Invert colors: QuickDraw library is black-on-white, 
            # but standard ML datasets (and your code) expect white-on-black (0-255)
            data = 255 - np.array(img_small)
            bitmaps.append(data.flatten())
            
        np.save(save_path, np.array(bitmaps))
    else:
        print(f"{category} already exists.")
```
